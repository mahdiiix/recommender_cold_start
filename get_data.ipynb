```python
import pandas as pd
import tweepy
import pickle
```

```python
auth = tweepy.AppAuthHandler(consumer_key= '...', consumer_secret= '...')
```

```python
api = tweepy.API(auth)
```

```python
def get_all_tweets_time(user_id):
    tweets = api.user_timeline(user_id=user_id, count=500)
    all_tweets_time = []
    all_tweets_time.extend([t.created_at for t in tweets])
    oldest = tweets[-1].id
    while True:
        tweets = api.user_timeline(user_id=user_id, count=500, max_id=oldest-1)
        if len(tweets) == 0:
            break
        all_tweets_time.extend([t.created_at for t in tweets])
        oldest = tweets[-1].id
    return all_tweets_time
```

```python
def get_user_features(user_id):
    user = api.get_user(user_id)
    return {'created_at': user.created_at,
            'friends_count': user.friends_count,
            'followers_count': user.followers_count,
            'favourites_count': user.favourites_count,
            'geo_enabled': user.geo_enabled,
            'statuses_count': user.statuses_count}
```

```python
users = pd.read_csv('users.dat', delimiter='::', names=['user_id', 'twitter_id'])
```

```python
all_feats = {}
errors = []
i = 0
for u in users.itertuples():
    try:
        all_feats[u.user_id] = (get_all_tweets_time(u.twitter_id), get_user_features(u.twitter_id))
        i += 1
        print(u.Index)
    except:
        errors.append(u.Index)
    if i == 250:
        with open(f'files/{u.Index}', 'wb') as f:
            pickle.dump(all_feats, f, protocol=pickle.HIGHEST_PROTOCOL)
        all_feats = {}
        i = 0
        
with open(f'{u.Index}', 'wb') as f:
    pickle.dump(all_feats, f, protocol=pickle.HIGHEST_PROTOCOL)
```
